```python
import pandas as pd
import numpy as np
```

Leo los datos.

```python
raw = pd.read_csv(".\\data\\grandslams.csv")

print(len(raw))
raw.head()
```

Quito 4 filas mal extraídas , además de crear la columna ganador y modificar la columna retirado

```python
filtro = raw['setsA'].isin(['Fils A.', 'Gaston H.', 'Djokovic N.', 'De Minaur A.'])
datos = raw[~filtro].copy()
datos = datos[datos['cuotaA'] != 'Show more']
datos['setsA'] = datos['setsA'].astype(int)
datos['setsB'] = datos['setsB'].astype(int)
datos['acesA'] = datos['acesA'].astype(int)
datos['acesB'] = datos['acesB'].astype(int)
conditions = [(datos["setsA"] == 3),(datos["setsB"] == 3)]
choices = [0, 1]
datos["ganador"] = np.select(conditions, choices, default=-1)
datos.loc[datos['ganador'] == -1, 'retirado'] = 1
datos.loc[datos['ganador'] != -1, 'retirado'] = 0

datos.head()
```

Observamos que hay bastantes columnas que solo contienen NaN

```python
print(datos.isna().sum()[datos.isna().sum() > 0])
```

Procedemos a eliminarlas

```python
nan_counts = datos.isna().sum()
cols_to_keep = nan_counts[nan_counts <= 500].index
datos = datos[cols_to_keep]

print(datos.isna().sum()[datos.isna().sum() > 0]) #también observamos que las columnas relacionadas con la velocidad del saque tienen muchos NaN pero no las eliminamos todavía
```

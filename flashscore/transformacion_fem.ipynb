```python
import pandas as pd
import numpy as np
```

Funciones

```python
def separar_df(df):
    filtro = df['torneo'].str.contains('Men|ATP', regex=True, case = False)
    fem = df[~filtro]
    return fem
```

Lectura

```python
raw = pd.read_csv(".\\data\\partidos_tenis_2025-06-11.csv")
fem1 = separar_df(raw)
fem2 = pd.read_csv(".\\data\\ilkley_grado_valencia_fem_2025-06-11.csv")
fem3 = pd.read_csv(".\\data\\biarritz-ceskalipa_fem_2025-06-11.csv")

dfs_fem = [fem1, fem2, fem3]
fem = pd.concat(dfs_fem, ignore_index=True)

fem
```
